```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px
import streamlit as st
```

```python
vehicles= pd.read_csv('vehicles_us.csv')
vehicles.info(show_counts=True)
vehicles.head(20)
```

```python
vehicles['Make'] = vehicles['model'].str.split().str[0]
vehicles['Model'] = vehicles['model'].str.split().str[~0]
```

```python
vehicles['date_posted'] = pd.to_datetime(vehicles['date_posted'], format='%Y-%m-%d')
vehicles['paint_color'] = vehicles['paint_color'].fillna('NaN')
vehicles['cylinders'] = vehicles['cylinders'].fillna('0000').astype(int)
```

```python

vehicles['model_year'] = vehicles['model_year'].fillna('0000').astype(int)
```
